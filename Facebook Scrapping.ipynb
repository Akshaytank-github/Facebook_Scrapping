```python
from tkinter import *
from tkinter.ttk import *
from tkinter.scrolledtext import ScrolledText
import pandas as pd
from facebook_scraper import get_posts
```

```python
root = Tk()
root.title('Facebook Scrapping')
root.geometry('1360x780')

def varistate():
    data_list = []

    for post in get_posts(account=namevalue.get(), pages=1):
        data_list.append(post)

    post_dataframe = pd.DataFrame(data_list)
    post_dataframe['time'] = pd.to_datetime(post_dataframe.time)

    mask = (post_dataframe['time'] <= enddatevalue.get()) & (post_dataframe['time'] >= stadatevalue.get())
    post_dataframe = post_dataframe.loc[mask]

    variable_list = []
    variable1 = int(var1.get())
    variable2 = int(var2.get())
    variable3 = int(var3.get())
    variable4 = int(var4.get())
    variable5 = int(var5.get())
    variable6 = int(var6.get())
    variable7 = int(var7.get())
    variable8 = int(var8.get())
    variable_list.append(variable1)
    variable_list.append(variable2)
    variable_list.append(variable3)
    variable_list.append(variable4)
    variable_list.append(variable5)
    variable_list.append(variable6)
    variable_list.append(variable7)
    variable_list.append(variable8)

    index_pos_list = [i for i in range(len(variable_list)) if variable_list[i] == 1]

    for ele in index_pos_list:
        if ele == 0:
            posttext_list = post_dataframe['post_text'].values.tolist()

            data = ''
            for i in posttext_list:
                data += i
                data += '\n\n'

            pote = Label(root, text='Post Text :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=10, y=300)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=210, y=300)
            editArea.insert(INSERT, data)

        if ele == 1:
            likes_list = post_dataframe['likes'].values.tolist()
            nol = Label(root, text='Number of Likes :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=10, y=400)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=210, y=400)
            editArea.insert(INSERT, likes_list)

        if ele == 2:
            share_list = post_dataframe['shares'].values.tolist()
            nos = Label(root, text='Number of Shares :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=10, y=500)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=210, y=500)
            editArea.insert(INSERT, share_list)

        if ele == 3:
            comment_list = post_dataframe['comments'].values.tolist()
            noc = Label(root, text='Number of Comments :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=10, y=600)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=210, y=600)
            editArea.insert(INSERT, comment_list)

        if ele == 4:
            postid_list = post_dataframe['post_id'].values.tolist()
            data = ''
            for i in postid_list:
                data += i
                data += '\n\n'
            pid = Label(root, text='Post ID :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=715, y=300)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=890, y=300)
            editArea.insert(INSERT, data)

        if ele == 5:
            posturl_list = post_dataframe['post_url'].values.tolist()
            data = ''
            for i in posturl_list:
                data += i
                data += '\n\n'
            purl = Label(root, text='Post URL :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=715, y=400)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=890, y=400)
            editArea.insert(INSERT, data)

        if ele == 6:
            time_list = post_dataframe['time']
            noc = Label(root, text='Post Date and Time :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=715, y=500)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=890, y=500)
            editArea.insert(INSERT, time_list)

        if ele == 7:
            image_list = post_dataframe['image'].values.tolist()
            purl = Label(root, text='Image link :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=715, y=600)
            editArea = ScrolledText(master=root, wrap=WORD, width=50, height=5, font='comicsansms 11')
            editArea.place(x=890, y=600)
            editArea.insert(INSERT, image_list)
            

head = Label(root, text='Welcome to Facebook Scrapping', font='comicsansms 19 bold', background='#4267B2', foreground='white')
head.place(x=480, y=12.5)

# Text for form
name = Label(root, text='Enter a valid Facebook page name :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(x=10,
                                                                                                                y=60)
start_date = Label(root, text='Enter starting date (YYYY-MM-DD format only) :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(
    x=10, y=90)
end_date = Label(root, text='Enter ending date (YYYY-MM-DD format only) :', font='comicsansms 11 bold', background='#4267B2', foreground='white').place(
    x=10, y=120)
non1 = Label(root, text=None).place(x=10, y=140)

# Variable for storing values-input
namevalue = StringVar()
stadatevalue = StringVar()
enddatevalue = StringVar()

# Variable for output
fbpostext = StringVar()

# Tkinter for entry
nameentry = Entry(root, textvariable=namevalue, font='comicsansms 11').place(x=500, y=60)
stadateentry = Entry(root, textvariable=stadatevalue, font='comicsansms 11').place(x=500, y=90)
enddateentry = Entry(root, textvariable=enddatevalue, font='comicsansms 11').place(x=500, y=120)

# Checkbox and Button styling
style = Style()
style.configure('W.TButton', font='comicsansms 11 bold', background='#4267B2')
style.configure('TCheckbutton', font='comicsansms 11', background='#4267B2', foreground='white')

# CheckBox and Button
line1 = Label(
    text='------ Please select from the following -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------',
    font='comicsansms 10 bold', background='#4267B2', foreground='white')
line1.place(x=0, y=160)

var1 = IntVar()
var2 = IntVar()
var3 = IntVar()
var4 = IntVar()
var5 = IntVar()
var6 = IntVar()
var7 = IntVar()
var8 = IntVar()

chkbtn1 = Checkbutton(root, text='Post_Text', takefocus=1, style='TCheckbutton', variable=var1).place(x=45, y=190)
chkbtn2 = Checkbutton(root, text='Number of Likes', takefocus=1, style='TCheckbutton', variable=var2).place(x=200,
                                                                                                            y=190)
chkbtn3 = Checkbutton(root, text='Number of Shares', takefocus=1, style='TCheckbutton', variable=var3).place(x=390,
                                                                                                             y=190)
chkbtn4 = Checkbutton(root, text='Number of Comments', takefocus=1, style='TCheckbutton', variable=var4).place(x=590,
                                                                                                               y=190)
chkbtn5 = Checkbutton(root, text='Post_ID', takefocus=1, style='TCheckbutton', variable=var5).place(x=820, y=190)
chkbtn6 = Checkbutton(root, text='Post_url', takefocus=1, style='TCheckbutton', variable=var6).place(x=960, y=190)
chkbtn7 = Checkbutton(root, text='Time', takefocus=1, style='TCheckbutton', variable=var7).place(x=1100, y=190)
chkbtn8 = Checkbutton(root, text='Image_url', takefocus=1, style='TCheckbutton', variable=var8).place(x=1210, y=190)

b1 = Button(text="Get Results", command=varistate, style='W.TButton').place(x=45, y=230)
line2 = Label(
    text='-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------',
    font='comicsansms 10 bold', background='#4267B2', foreground='white')
line2.place(x=0, y=270)
root['bg'] = '#4267B2'

root.mainloop()
```
